```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Unemployment in India.csv")
```

```python
df.head()
print("\nDataset Info:")
df.info()
```

```python
df.columns=df.columns.str.strip()
```

```python
print("\Summary Statistics:")
print(df.describe(include="all"))
```

```python
# Visualization Section

plt.figure(figsize=(10,6))
sns.histplot(df['Estimated Unemployment Rate (%)'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Unemployment Rate in India", fontsize=14)
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Average Unemployment by Region

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Region', y='Estimated Unemployment Rate (%)', ci=None, palette='viridis')
plt.xticks(rotation=45)
plt.title("Average Unemployment Rate by Region", fontsize=14)
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Time-series trend (if Date column exists)

if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date'])
    plt.figure(figsize=(14,6))
    sns.lineplot(data=df, x='Date', y='Estimated Unemployment Rate (%)', hue='Region')
    plt.title("Unemployment Trend Over Time", fontsize=14)
    plt.ylabel("Unemployment Rate (%)")
    plt.show()
```

```python
# Correlation heatmap

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplot for regional comparison

plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Region', y='Estimated Unemployment Rate (%)', palette='Set2')
plt.xticks(rotation=45)
plt.title("Unemployment Rate Distribution by Region")
plt.show()
```
